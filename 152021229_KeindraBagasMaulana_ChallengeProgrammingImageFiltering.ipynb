Low Pass filter

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def apply_low_pass_filter(image, kernel_size):
    # Membuat kernel low pass filter
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    # Menggunakan fungsi filter2D dari OpenCV untuk menerapkan kernel pada citra
    filtered_image = cv2.filter2D(image, -1, kernel)
    return filtered_image

# Membaca gambar input dalam mode grayscale
input_image = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

# Memanggil fungsi apply_low_pass_filter untuk melakukan filtering
lpf_filtered_image = apply_low_pass_filter(input_image, kernel_size=5)

# Menampilkan gambar hasil filtering
cv2_imshow(lpf_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

High Pass

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def apply_high_pass_filter(image, kernel_size):
    # Membuat kernel high pass filter
    kernel = np.ones((kernel_size, kernel_size), np.float32) * -1
    center = kernel_size // 2
    kernel[center, center] = kernel_size * kernel_size - 1
    # Menggunakan fungsi filter2D dari OpenCV untuk menerapkan kernel pada citra
    filtered_image = cv2.filter2D(image, -1, kernel)
    return filtered_image

# Membaca gambar input dalam mode grayscale
input_image = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

# Memanggil fungsi apply_high_pass_filter untuk melakukan filtering
hpf_filtered_image = apply_high_pass_filter(input_image, kernel_size=3)

# Menampilkan gambar hasil filtering
cv2_imshow(hpf_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

Kernel sobel

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Membaca gambar input dalam mode grayscale
input_image = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

# Membuat kernel Sobel untuk deteksi tepi horizontal dan vertikal
kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32)

# Melakukan konvolusi dengan kernel Sobel
output_image_x = cv2.filter2D(input_image, -1, kernel_x)
output_image_y = cv2.filter2D(input_image, -1, kernel_y)

# Menggabungkan hasil deteksi tepi horizontal dan vertikal
output_image = cv2.addWeighted(output_image_x, 0.5, output_image_y, 0.5, 0)

# Menampilkan gambar hasil filtering
cv2_imshow(output_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
